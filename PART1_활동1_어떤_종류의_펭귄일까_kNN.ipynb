# 원-핫 인코딩하기

```python
penguins_encoding = pd.get_dummies(penguins, columns=['island', 'sex'])  #명목형 변수를 원핫인코딩
```

```python
penguins_encoding.head()
```

# 독립 변수와 종속 변수 구분하기

```python
X = penguins_encoding.drop(['species'], axis = 1)  #독립변수
y = penguins_encoding['species']                    #종속변수
```

# 훈련 데이터와 테스트 데이터로 분리하기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 11) #훈련용 독립변수, 테스트용 독립변수, 훈련용 종속변수, 테스트용 종속변수로 분할
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)  #분할한 데이터 셋의 형태 출력
```

# kNN 모델 생성, 학습하기

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
knn = KNeighborsClassifier(n_neighbors=k)  #모델 생성
knn.fit(X_train, y_train)  #모델 학습
```

# 훈련용 데이터로 학습한 평가 지표 확인하기

```python
print(knn.score(X_train, y_train))
```

# 테스트하기

```python
y_pred = knn.predict(X_test)  # y_pred: 테스트용 독립 변수에 대해 예측한 값
y_pred[:20]
```

# 평가하기

```python
print('knn accuracy:{:.3f}'.format(knn.score(X_test, y_test)))
```

# k값에 따른 성능 비교하기

```python
import numpy as np
import matplotlib.pyplot as plt
k_range = range(1, 31)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.plot(k_range, scores)
plt.xlabel('value of K')
plt.ylabel('Accuracy Score')
plt.xticks(np.arange(0, 31, 5))
plt.grid()
plt.title('k-NN Accuracy score')
plt.show()
```

```python
for value, score in enumerate(scores):
    print('k={}일 때: accuracy:{:.3f}'.format(value + 1, score))
```

# 6. 모델성능 개선하기

# 훈련 데이터 정규화하기

```python
from sklearn.preprocessing import MinMaxScaler  #정규화
scaler = MinMaxScaler()
scaler.fit(X)
scaled = scaler.transform(X)
round(pd.DataFrame(scaled).describe(), 2)
```

# 훈련용 데이터와 테스트용 데이터로 분리하기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size = 0.3, random_state = 11)  #정규화한 scaled 변수를 사용
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# 모델 생성 및 학습하기

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
print(knn.score(X_train, y_train))
```

# 테스트 및 평가하기

```python
y_pred = knn.predict(X_test)
y_pred[:20]
```

```python
print('knn accuracy:{:.3f}'.format(knn.score(X_test, y_test)))
```

# k값 정하기

```python
import numpy as np
import matplotlib.pyplot as plt
k_range = range(1, 31)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.plot(k_range, scores)
plt.xlabel('value of K')
plt.ylabel('Accuracy Score')
plt.xticks(np.arange(0, 31, 5))
plt.grid()
plt.title('k-NN Accuracy score')
plt.show()
```

# 해 보기: k값에 따른 score 변수값 출력하기

```python
for value, score in enumerate(scores):
    print('k={}일 때: accuracy:{:.3f}'.format(value+1, score))
```

# 스케일링 두 가지 방법

```python
#최소 최대 정규화
import numpy as np
height = [100, 120, 160, 180]
weight = [20, 30, 50, 70]
print(np.min(height), np.max(height))
print(np.min(weight), np.max(weight))

for x in height:
  print((x - np.min(height)) / (np.max(height) - np.min(height)), end=' ')
print()

for x in weight:
  print((x - np.min(weight)) / (np.max(weight) - np.min(weight)), end=' ')
```

```python
#표준화 방법
import numpy as np
height = [100, 120, 160, 180]
weight = [20, 30, 50, 70]
print(np.mean(height), round(np.std(height), 1))
print(np.mean(weight), round(np.std(weight), 1))

for i in height:
  print(round((i-np.mean(height))/np.std(height), 2), end=' ')
print()

for i in weight:
  print(round((i-np.mean(weight))/np.std(weight), 2), end=' ')
```
